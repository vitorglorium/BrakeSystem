```python
def ufprformula_fc():
  ##fonte é funky_dr | https://pt.rakko.tools/tools/68/
    print("==========================================================================================================")
    print('##   ##   ######   #####    #####             ######   #####    #####    ######  ##   ##  ##        ##### ')
    print('##   ##  ##       ##   ##  ##   ##           ##       ##   ##  ##   ##  ## ## ## ##   ##  ##       ##   ##')
    print('##   ##  ## ###   ## ###   ## ###            ## ###   ##   ##  ## ###   ## ## ## ##   ##  ##       ## ####')
    print('##   ##  ##       ##       ##   ##           ##       ##   ##  ##   ##  ## ## ## ##   ##  ###      ##   ##')
    print(' #####   ##       ##       ##   ##           ##        #####   ##   ##  ## ## ##  #####    ######  ##   ##')
    print("========================================== FREIOS E COCKPIT ==============================================")
    print("                Este modelo foi criado com base em pedais do tipo fixo com CMs fixos.                     ")
    print("                Para manter conformidade nos resultados, use SEMPRE unidades padrões.                     ")
    print("                      Metros | Segundos | Pascal (N/m²) | Newtons | Kg | etc.                             ")
    print("                                        By: Vitor Glorium                                                 ")
    print("==========================================================================================================")
    print("            A vida sempre recompensa os que tentam, não importa como, apenas tente, apenas comece.        ")
    print("==========================================================================================================")

def instrucoes(): #instrucoes para o uso
    print("Se você NÃO tiver a informação, digite 'N/A' ")

def dicionario(): #esta função tem como objetivo mostrar o significado de todas as variaveis quando salvas para pleno entendimento
    print("----------------------------------------------------------------------------------------------------------")
    pass

# -------------------- Funções De Dimensionamento  ----------------------------------
def ganho_mecanico(parametros_globais): #Ganho Mecânico do Pedal
    lb = parametros_globais["lb"]
    lp = parametros_globais["lp"]
    gm = lp/lb
    outputs["gm"]=gm

def fbar(parametros_globais,outputs): #Força Gerada na Balance Bar
    fpe = parametros_globais["fpe"]
    gm = outputs["gm"]
    fbias = gm*fpe
    outputs["fbias"]=fbias

def desloc_bar(parametros_globais,outputs): #Calcula o Deslocamento da Balance Bar - (DIANTEIRO E TRASEIRO)
    gm = outputs["gm"]
    xp = parametros_globais["xp"]
    xbias = xp/gm
    outputs["xbias"]=xbias

def bias(parametros_globais,outputs): #calcula a relação de BIAS do carro - (DIANTEIRO E TRASEIRO)
    yr = parametros_globais["yr"]
    yf = parametros_globais["yf"]
    yb = yr + yf #Tamanho útil da balance bar
    xbf = yr/yb
    xbr = 1-xbf
    bias_front = round(xbf * 100)
    bias_rear = round(xbr * 100)
    BIAS = f"{bias_front}%/{bias_rear}% (FRONT/REAR)"
    parametros_globais["yb"]=yb
    outputs["BIAS"]=BIAS #aqui ele vai retornar o bias do carro no formato "XX%/YY% (FRONT/REAR)"
    outputs["xbf"]=xbf
    outputs["xbr"]=xbr

def fm(parametros_globais,outputs): #Calcula as forças geradas no cilindro mestre Front (Fmf) | Rear (Fmr)
    xbf = outputs["xbf"]
    xbr = outputs["xbr"]
    fbias = outputs["fbias"]
    fcmf = (xbf*fbias)
    fcmr = (xbr*fbias)
    outputs["fcmf"]=fcmf
    outputs["fcmr"]=fcmr

def angs(parametros_globais,outputs): #Calcular o ang desloc pelo pedal de Freio (e do BIA no ref do pedal), calcular tbm pensando na haste da balance bar o angulado para futuramente trabalhar c/ isso.
    yb = parametros_globais["yb"]
    xmf = parametros_globais["xmf"]
    xmr = parametros_globais["xmr"]
    xbias = outputs["xbias"]
    xbf = outputs["xbf"]

    vetor_neutro = np.array([1,1,1])                 #Vetor neutro usado para contas, neutro pela visão matricial de identidade.
    vetor_posf = np.array([xmf,xbias,xmr])           #Vetorno formato ([Deslocamento do CM Front, Deslocamento da Balance Bar, Deslocamento do CM Rear])
    vetor_subtra = vetor_neutro*(np.min(vetor_posf)) #Vetor a ser subtraido do posf para encontrarmos o angulo
    vetor_final = (vetor_posf)-(vetor_subtra)        #Vetor do deslocamento final relativo DFR ([CFR do CM Front, CFR da Balance Bar, CFR do CM Rear ])
    maximo = np.max(vetor_final)                     #É o nosso cateto oposto de maior valor, usado para que junto do yb (cte) achemos o angulo variado pela balance bar quando em operação

    tangente = maximo / yb
    angulo_graus = math.degrees(math.atan(tangente))
    outputs["angulo_bbar"]=angulo_graus

    lb = parametros_globais["lb"]
    tangente_pedal = xbias / lb
    angulo_graus2 = math.degrees(math.atan(tangente_pedal))
    outputs["angulo_pedal"]=angulo_graus2

def pressao_cm(parametros_globais,outputs): #Calcula a pressão gerada nos cilindros mestres, cmf (front) e cmr (rear) - (DIANTEIRO E TRASEIRO)
    fcmf = outputs["fcmf"]
    fcmr = outputs["fcmr"]
    acmf = parametros_globais["acmf"]
    acmr = parametros_globais["acmr"]
    pf = fcmf/acmf
    pr = fcmr/acmr
    outputs["pf"] = pf
    outputs["pr"] = pr

def forca_pistao_pinca(parametros_globais,outputs):
    pf = outputs["pf"]
    pr = outputs["pr"]

    apf = parametros_globais["apf"]
    apr = parametros_globais["apr"]

    fpistf = (apf)*(pf)
    fpistr = (apr)*(pr)

    outputs["fpistf"]=fpistf #Força gerada por cada pistão na pinça dianteira
    outputs["fpistr"]=fpistr #Força gerada por cada pistão na pinça traseira

def forca_atrito(parametros_globais,outputs): #Calcula a força de atrito gerada durante a frenagem - (DIANTEIRO E TRASEIRO)
    npf = parametros_globais["npf"] # N° de pistões DIANTEIRO por pinça
    npr = parametros_globais["npr"] # N° de pistões TRASEIRO por pinça

    fpistf = outputs["fpistf"] # Força gerada por pistão na DIANTEIRA
    fpistr = outputs["fpistr"] # Força gerada por pistão na TRASEIRA

    udpf = parametros_globais["udpf"]
    udpr = parametros_globais["udpr"]

    fatf = 2*fpistf*udpf*npf # Força de atrito gerada POR PINÇA - NA DIANTEIRA (POR PINÇA/RODA!)  #OBSERVAÇÃO AQUI POIS TEM QUE REFLETIR BEM, DIFERE DO MATERIAL DO ERTHAL, FIZ PARA TORNAR AMPLO.
    fatr = 2*fpistr*udpr*npr # Força de atrito gerada POR PINÇA - NA TRASEIRA (POR PINÇA/RODA!) #OBSERVAÇÃO AQUI POIS TEM QUE REFLETIR BEM, DIFERE DO MATERIAL DO ERTHAL, FIZ PARA TORNAR AMPLO.

    outputs["fatf"]=fatf
    outputs["fatr"]=fatr

def torque(parametros_globais,outputs): #Calcula o torque de frenagem gerado na roda - (DIANTEIRO E TRASEIRO)

    fatf = outputs["fatf"]
    fatr = outputs["fatr"]

    ddf = parametros_globais["ddf"]
    ddr = parametros_globais["ddr"]

    dof = parametros_globais["dof"]
    dor = parametros_globais["dor"]

    raiof = ddf/2
    raior = ddr/2

    raioefetivof = raiof-dof
    raioefetivor = raior=dor

    torquef = (fatf)*(raioefetivof)
    torquer = (fatr)*(raioefetivor)

    outputs["torquef"] = torquef
    outputs["torquer"] = torquer


def coef_pinca(parametros_globais,outputs): #calcula o coeficiente de pinça - (DIANTEIRO E TRASEIRO)

  # O coeficiente de pinça que o Erthal calcula no livro é uma denominação conveninete para trabalhar com Cf ou Cr nas formulas, porém, tem outra tratativa a qual vou usar aqui.
  # Basicamente trato coeficiente de pinça C como "um rendimento/grau de eficiencia" em converter a normal em atrito de fato.
  # C = (Força de Atrito) / (Força Normal)

# --------------------- Coeficiente by Minha Metodologia -------------------------
    fatf = outputs["fatf"]
    fatr = outputs["fatr"]

    fpistf = outputs["fpistf"]   #Força gerada por cada pistão na pinça dianteira
    fpistr = outputs["fpistr"]   #Força gerada por cada pistão na pinça traseira

    npf = parametros_globais["npf"] # N° de pistões DIANTEIRO por pinça  -------> Estou normalizando a analise com isso.
    npr = parametros_globais["npr"] # N° de pistões TRASEIRO por pinça   -------> Estou normalizando a analise com isso.

    forcNf= (npf)*(fpistf)   # -----> Como a força de atrito já esta proporcional ao N° de pistões, coloquei a fornça normal TAMBÉM proprocional ao N° de pistões p/ fazer o comparativo vide "C"
    forcNr = (npr)*(fpistr)  # -----> Como a força de atrito já esta proporcional ao N° de pistões, coloquei a fornça normal TAMBÉM proprocional ao N° de pistões p/ fazer o comparativo vide "C"

    CoefPinF = (fatf)/(forcNf)
    CoefPinR = (fatr)/(forcNr)

    outputs["CoefPinF"] = CoefPinF
    outputs["CoefPinR"] = CoefPinR
# --------------------- Coeficiente by Erthal's Metodologia -----------------------

    Cff = 1
    Cfr = 1

    # Valor Temporário

    outputs["Cff"] = Cff
    outputs["Cfr"] = Cfr


   ##CoefR = 2 * Area do pistão da pínça dianteira * numero de pistoes








def volume_desloc_pinca(parametros_globais,outputs): #fornece um total de volume deslocado (Pinça) - (DIANTEIRO E TRASEIRO)
    xppf = parametros_globais["xppf"]
    xppr = parametros_globais["xppr"]

    npf = parametros_globais["npf"]
    npr = parametros_globais["npr"]

    apf = parametros_globais["apf"]
    apr = parametros_globais["apr"]

    dVpf = (xppf)*(npf)*(apf)  #Volume deslocado na pinça DIANTEIRA
    dVpr = (xppr)*(npr)*(apr)  #Volume deslocado na pinça TRASEIRA

    outputs["dVPf"] = dVpf
    outputs["dVPr"] = dVpr

  #Esse volume é referente a UMA UNIDADE DE PINÇA DE FREIO
  # REVER ESSA MERDA DO CARALHO!

def volume_total_fluido(parametros_globais,outputs): #fornece um total de volume deslocado (Cilindro mestre) - (DIANTEIRO E TRASEIRO)
    xlff = parametros_globais["xlff"]
    xlfr = parametros_globais["xlfr"]
    vpuml = parametros_globais["vpumlf"]
    npf = parametros_globais["npf"]
    npr = parametros_globais["npr"]

    #Volume das Linhas
    volumeLinhaF = xlff*vpuml #Volume total na linha dianteira
    volumeLinhaR = xlfr*vpuml #Volume total na linha traseira

    #Volume da Pinça
    VoolFluPistaoF = parametros_globais["VoolFluPistaoF"]
    VoolFluPistaoR = parametros_globais["VoolFluPistaoR"]

    VoolFluPistaoFtot = VoolFluPistaoF*2*npf # (Volume por Pistão) X (Número de Pistões por Pinça) x 2 (Duas pinças por sistema) ->Volume Total de Fluído nas Pinças
    VoolFluPistaoRtot = VoolFluPistaoR*2*npr # (Volume por Pistão) X (Número de Pistões por Pinça) x 2 (Duas pinças por sistema) ->Volume Total de Fluído nas Pinças

    # Sujestão: Aqui um ponto de melhoria seria usar um volume único por pinça, porém como a diferença é pouca vou usar por pistão.

    # Volume no Cilindro Mestre
    VolumeCMF = parametros_globais["VolumeCMF"]
    VolumeCMR = parametros_globais["VolumeCMR"]


    VtotF = VolumeCMF+VoolFluPistaoFtot+volumeLinhaF
    VtotR = VolumeCMR+VoolFluPistaoRtot+volumeLinhaR

    outputs["VTotF"]=VtotF  # VOLUME TOTAL DO SISTEMA DIANTEIRO
    outputs["VTotR"]=VtotR  # VOLUME TOTAL DO SISTEMA DIANTEIRO


def volume_cm(parametros_globais,outputs): #fornece um total de volume deslocado (Cilindro mestre) - (DIANTEIRO E TRASEIRO)

  #Observação: Nas contas, diferente do Erthal, utilizei o volume TOTAL presente nos sistema de freios...
  # Ele toma como volume total apenas os volumes das pinças + CM, porém, EU, utilizo o volume total da linha.
  # A aproximação utilizada pelo Erthal é mais bem tratada quando observamos o curso do pedal ou deslocamento minimo do cilindro mestre (próxima função, deslocamento_min_cm())

  # Volume Deslocado é igual ao volume deslocado nas pinças mais o volume necessário devido ompressibilidade do fluido (função do volume total):

    B = parametros_globais["B"]

    acmf = parametros_globais["acmf"]
    acmr = parametros_globais["acmr"]

    xcmf = parametros_globais["xcmf"]
    xcmr = parametros_globais["xcmr"]

    vcmf = (acmf)*(xcmf)
    vcmr = (acmr)*(xcmr)

    pf = outputs["pf"]
    pr = outputs["pr"]

    dVpf = outputs["dVPf"]
    dVpr = outputs["dVPr"]

    VtotF = outputs["VTotF"]  # VOLUME TOTAL DO SISTEMA DIANTEIRO
    VtotR = outputs["VTotR"]  # VOLUME TOTAL DO SISTEMA DIANTEIRO

    dVpfTOT = dVpf*2 #Aqui é o volume total descloado no sistema DIANTEIRO
    dVprTOT = dVpr*2 #Aqui é o volume total descloado no sistema TRASEIRO


    VDeslocadoCMF = dVpfTOT + ((VtotF/B)*pf)
    VDeslocadoCMR = dVprTOT + ((VtotR/B)*pr)

    outputs["VDeslocadoCMF"] = VDeslocadoCMF
    outputs["VDeslocadoCMR"] = VDeslocadoCMR

def deslocamento_min_cm(parametros_globais,outputs):

    VDeslocadoCMF = outputs["VDeslocadoCMF"]
    VDeslocadoCMR = outputs["VDeslocadoCMR"]

    acmf = parametros_globais["acmf"]
    acmr = parametros_globais["acmr"]

    dXminCMF = VDeslocadoCMF/acmf
    dXminCMR = VDeslocadoCMR/acmr

    outputs["dXminCMF"]=dXminCMF
    outputs["dXminCMR"]=dXminCMR

  # Esse valor é importante para saber se o Cilindro Mestre tem o mínimo de deslocamento.

def forc_frenagem(parametros_globais,outputs): #Calcula a força de frenagem atuando no carro - (DIANTEIRO E TRASEIRO)

    torquef = outputs["torquef"]
    torquer = outputs["torquer"]

    dtf = parametros_globais["dtf"]
    dtr = parametros_globais["dtr"]

    ups = parametros_globais["ups"]
    upm = parametros_globais["upm"]

    ff = torquef/(dtf/2)
    fr = torquer/(dtr/2)

    num_de_discos_frente = 2
    num_de_discos_tras = 2

    Ff = ff*num_de_discos_frente
    Fr = fr*num_de_discos_tras

    outputs["Ff"] = Ff
    outputs["Fr"] = Fr

def relac_forc_frenagem(parametros_globais,outputs):
    #A relação entre a força de frenagem e outros elemntos diversos como proposto pelo Erthal depende do coeficiente de pinça
    #NESTA situação usarei a mesma ideia de coeficiente de pinça proposta pelo Erthal.



    ###ESSA PARTE AQUI ESTOU ESPERANDO O ERTHAL ME AUXILIAR!
    ### VER NO COEFICIENTE DE PINÇA MAIS PRA CIMA (FUNÇÃO) CRIEI UM COEF. DE PINÇA IGUAL AO PROPOSTO PELO ERTHAL.



    pass

def peso_estatico(parametros_globais,outputs):
    g = 9,80665

    mcar1 = parametros_globais["mcar1"] # Somente o carro
    mcar2 = parametros_globais["mcar2"] # Carro com piloto e gasolina
    b = parametros_globais["b"] # Distancia entre os eixos do carro.
    Zf = parametros_globais["Zf"] # Distancia do C.G até o eixo dianteiro
    Zr = parametros_globais["Zr"] # Distancia do C.G até o eixo traseiro
    H = parametros_globais["H"] # Altura do chão até o C.G do carro
    a = parametros_globais["a"] # Desaceleração em módulo

    w = mcar1*g
    wt = mcar1*g
    wf_e = (Zr/b)*w  # Peso estático - Dianteiro [SECO]
    wr_e = (Zf/b)*w  # Peso estático - Traseiro [SECO]

    wf_e_t = (Zr/b)*wt
    wr_e_t = (Zf/b)*wt

  # Peso SECO (sem piloto e demais fluídos)
    outputs["wf_e"]= wf_e
    outputs["wr_e"]= wf_e

  # Peso TOTAL (com piloto e demais fluídos)
  outputs["wf_e_t"] = wf_e_t
  outputs["wr_e_t"] = wr_e_t


def peso_dinamico(parametros_globais,outputs):
    g = 9,80665

    mcar1 = parametros_globais["mcar1"] # Somente o carro
    mcar2 = parametros_globais["mcar2"] # Carro com piloto e gasolina
    b = parametros_globais["b"] # Distancia entre os eixos do carro.
    Zf = parametros_globais["Zf"] # Distancia do C.G até o eixo dianteiro
    Zr = parametros_globais["Zr"] # Distancia do C.G até o eixo traseiro
    H = parametros_globais["H"] # Altura do chão até o C.G do carro
    a = parametros_globais["a"] # Desaceleração em módulo
    w = marc*g

    G = a/g # normalizando a desaceleração em "g's"

    wf_e = outputs["wf_e"]
    wr_e = outputs["wr_e"]

    wf_e_t = outputs["wf_e_t"]
    wr_e_t = outputs["wr_e_t"]


    #PESO SECO
    Wf_d_seco = w*((wf_e)+((H/b)*G))  # Peso em função da desaceleração do carro - DIANTEIRA
    Wr_d_seco = w*((wr_e)-((H/b)*G))  # Peso em função da desaceleração do carro - TRASEIRA

    outputs["Wf_d_seco"] = Wf_d_seco
    outputs["Wr_d_seco"] = Wr_d_seco

    #PESO TOTAL
    Wf_d_total = w*((wf_e_t)+((H/b)*G))  # Peso em função da desaceleração do carro - DIANTEIRA
    Wr_d_total = w*((wr_e_t)-((H/b)*G))  # Peso em função da desaceleração do carro - TRASEIRA

    outputs["Wf_d_total"] = Wf_d_total
    outputs["Wr_d_total"] = Wr_d_total


def forca_ideal(parametros_globais,outputs):
  Cff = outputs["Cff"]
  Cfr = outputs["Cfr"]
  g = 9,80665
  w1 = mcar1*g # Peso somente do carro (seco)
  w2 = mcar2*g # Peso com piloto e gasolina
  mcar1 = parametros_globais["mcar1"] # Somente o carro (seco)
  mcar2 = parametros_globais["mcar2"] # Carro com piloto e gasolina
  mcar1 = parametros_globais["mcar1"] # Somente o carro
  mcar2 = parametros_globais["mcar2"] # Carro com piloto e gasolina
  b = parametros_globais["b"] # Distancia entre os eixos do carro.
  Zf = parametros_globais["Zf"] # Distancia do C.G até o eixo dianteiro
  Zr = parametros_globais["Zr"] # Distancia do C.G até o eixo traseiro
  H = parametros_globais["H"] # Altura do chão até o C.G do carro
  a = parametros_globais["a"] # Desaceleração em módulo
  ups = parametros_globais["ups"] # Atrito do pneu seco
  upm = parametros_globais["upm"] # Atrito do pneu molhado

  #PESO ESTÁTICO SECO
  wf_e = outputs["wf_e"] # Peso estatico FRONT - XBF
  wf_e = outputs["wr_e"] # Peso estatico REAR - XBR

  #PESO ESTATICO TOTAL
  wf_e_t = outputs["wf_e_t"] # Peso estático TOTAL - FRONT -XBF
  wr_e_t = outputs["wr_e_t"] # Peso estático TOTAL - REAR -XBF

# -------- Força Ideal | Pista Seca | Massa Seca ----------
  FI_FRONT_w1_SECO_s = w1(ups(wf_e+((H/b)*ups)))
  FI_REAR_w1_SECO_s = w1(ups(wr_e-((H/b)*ups)))
# -------- Força Ideal | Pista Seca | Massa Total  --------
  FI_FRONT_w2_SECO_t = w2(ups(wf_e_t+((H/b)*ups)))
  FI_REAR_w2_SECO_t = w2(ups(wr_e_t-((H/b)*ups)))

# -------- Força Ideal | Pista Molhada | Massa Seca -------
  FI_FRONT_w1_MOLHADO_s = w1(upm(wf_e+((H/b)*upm)))
  FI_REAR_w1_MOLHADO_s = w1(upm(wr_e-((H/b)*upm)))
# -------- Força Ideal | Pista Molhada | Massa Total ------
  FI_FRONT_w2_MOLHADO_t = w2(upm(wf_e_t+((H/b)*upm)))
  FI_REAR_w2_MOLHADO_t = w2(upm(wr_e_t-((H/b)*upm)))
# ------------------------------------------------------

  outputs["FI_FRONT_w1_SECO_s"]=FI_FRONT_w1_SECO_s
  outputs["FI_REAR_w1_SECO_s"]=FI_REAR_w1_SECO_s
  outputs["FI_FRONT_w2_SECO_t"]=FI_FRONT_w2_SECO_t
  outputs["FI_REAR_w2_SECO_t"]=FI_REAR_w2_SECO_t
  outputs["FI_FRONT_w1_MOLHADO_s"]=FI_FRONT_w1_MOLHADO_s
  outputs["FI_REAR_w1_MOLHADO_s"]=FI_REAR_w1_MOLHADO_s
  outputs["FI_FRONT_w2_MOLHADO_t"]=FI_FRONT_w2_MOLHADO_t
  outputs["FI_REAR_w2_MOLHADO_t"]=FI_REAR_w2_MOLHADO_t


def pressao_ideal(parametros_globais,outputs):

    #Teremos dois cases:
    # Case 1: Pista Seca
    # Case 2: Pista molhada
    # Retornar os dois pares (Pf_seco,Pr_seco) e (Pf_molhado,Pr_molhado)
  Cff = outputs["Cff"]
  Cfr = outputs["Cfr"]
  g = 9,80665
  w1 = mcar1*g # Peso somente do carro (seco)
  w2 = mcar2*g # Peso com pilto e gasolina
  mcar1 = parametros_globais["mcar1"] # Somente o carro (seco)
  mcar2 = parametros_globais["mcar2"] # Carro com piloto e gasolina
  b = parametros_globais["b"] # Distancia entre os eixos do carro.
  Zf = parametros_globais["Zf"] # Distancia do C.G até o eixo dianteiro
  Zr = parametros_globais["Zr"] # Distancia do C.G até o eixo traseiro
  H = parametros_globais["H"] # Altura do chão até o C.G do carro
  a = parametros_globais["a"] # Desaceleração em módulo
  ups = parametros_globais["ups"]
  upm = parametros_globais["upm"]
  dtf = parametros_globais["dtf"]
  dtr = parametros_globais["dtr"]
  ddf = parametros_globais["ddf"]
  ddr = parametros_globais["ddr"]
  dof = parametros_globais["dof"]
  dor = parametros_globais["dor"]
  nf = 2 # Número de Discos de Freio - FRONT
  nr = 2 # Número de Discos de Freio -

  #PESO ESTÁTICO SECO
  wf_e = outputs["wf_e"] # Peso estatico FRONT - XBF
  wf_e = outputs["wr_e"] # Peso estatico REAR - XBR

  #PESO ESTATICO TOTAL
  wf_e_t = outputs["wf_e_t"] # Peso estático TOTAL - FRONT -XBF
  wr_e_t = outputs["wr_e_t"] # Peso estático TOTAL - REAR -XBF

# -------- Pressão Ideal | Pista Seca | Massa Seca ----------
  PI_FRONT_SECO_w1_s = (dtf*w1*ups(wf_e+((H/b)*ups))) / (2*nf*cff*((ddf/2)-dof))
  PI_REAR_SECO_w1_s = (dtr*w1*ups(wr_e-((H/b)*ups))) / (2*nr*cfr*((ddr/2)-dor))

# -------- Pressão Ideal | Pista Seca | Massa Total  --------
  PI_FRONT_SECO_w2_t = (dtf*w2*ups(wf_e_t+((H/b)*ups))) / (2*nf*cff*((ddf/2)-dof))
  PI_REAR_SECO_w2_t = (dtr*w2*ups(wr_e_t-((H/b)*ups))) / (2*nr*cfr*((ddr/2)-dor))

# -------- Pressão Ideal | Pista Molhada | Massa Seca -------
  PI_FRONT_MOLHADO_w1_s = (dtf*w1*upm(wf_e+((H/b)*upm))) / (2*nf*cff*((ddf/2)-dof))
  PI_REAR_MOLHADO_w1_s = (dtf*w1*upm(wr_e+((H/b)*upm))) / (2*nf*cff*((ddf/2)-dof))

# -------- Pressão Ideal | Pista Molhada | Massa Total ------
  PI_FRONT_MOLHADO_w2_t = (dtf*w2*upm(wf_e_t+((H/b)*upm))) / (2*nf*cff*((ddf/2)-dof))
  PI_REAR_MOLHADO_w2_t = (dtf*w2*upm(wr_e_t+((H/b)*upm))) / (2*nf*cff*((ddf/2)-dof))

  outputs["PI_FRONT_SECO_w1_s"]=PI_FRONT_SECO_w1_s
  outputs["PI_REAR_SECO_w1_s"]=PI_REAR_SECO_w1_s
  outputs["PI_FRONT_SECO_w2_t"]=PI_FRONT_SECO_w2_t
  outputs["PI_REAR_SECO_w2_t"]=PI_REAR_SECO_w2_t
  outputs["PI_FRONT_MOLHADO_w1_s"]=PI_FRONT_MOLHADO_w1_s
  outputs["PI_REAR_MOLHADO_w1_s"]=PI_REAR_MOLHADO_w1_s
  outputs["PI_FRONT_MOLHADO_w2_t"]=PI_FRONT_MOLHADO_w2_t
  outputs["PI_REAR_MOLHADO_w2_t"]=PI_REAR_MOLHADO_w2_t


def bias_ideal(parametros_globais,outputs):
    #Teremos dois cases:
    # Case 1: Pista Seca
    # Case 2: Pista molhada
    # Retornar os dois pares (Xf_seco,Xr_seco) e (Xf_molhado,Xr_molhado)
  Cff = outputs["Cff"]
  Cfr = outputs["Cfr"]
  g = 9,80665
  w1 = mcar1*g # Peso somente do carro (seco)
  w2 = mcar2*g # Peso com pilto e gasolina
  mcar1 = parametros_globais["mcar1"] # Somente o carro (seco)
  mcar2 = parametros_globais["mcar2"] # Carro com piloto e gasolina
  b = parametros_globais["b"] # Distancia entre os eixos do carro.
  Zf = parametros_globais["Zf"] # Distancia do C.G até o eixo dianteiro
  Zr = parametros_globais["Zr"] # Distancia do C.G até o eixo traseiro
  H = parametros_globais["H"] # Altura do chão até o C.G do carro
  a = parametros_globais["a"] # Desaceleração em módulo
  ups = parametros_globais["ups"]
  upm = parametros_globais["upm"]
  dtf = parametros_globais["dtf"]
  dtr = parametros_globais["dtr"]
  ddf = parametros_globais["ddf"]
  ddr = parametros_globais["ddr"]
  dof = parametros_globais["dof"]
  dor = parametros_globais["dor"]
  nf = 2 # Número de Discos de Freio - FRONT
  nr = 2 # Número de Discos de Freio - REAR

  #PESO ESTÁTICO SECO
  wf_e = outputs["wf_e"] # Peso estatico FRONT - XBF
  wf_e = outputs["wr_e"] # Peso estatico REAR - XBR

  #PESO ESTATICO TOTAL
  wf_e_t = outputs["wf_e_t"] # Peso estático TOTAL - FRONT -XBF
  wr_e_t = outputs["wr_e_t"] # Peso estático TOTAL - REAR -XBF


# -------------------- Funções Gráficas ----------------------------------

def relac_pressoes(parametros_globais,outputs): #formula 4.28 da relação linear das pressões

    ###ESSA PARTE AQUI ESTOU ESPERANDO O ERTHAL ME AUXILIAR!
    pass
def relacao_for_frenagem(parametros_globais,outputs): #formula 4.29 que realciona linearmente as forças dianteira e traseira do carro.
    ###ESSA PARTE AQUI ESTOU ESPERANDO O ERTHAL ME AUXILIAR!
    pass

def desaceleracoes_feat_pesos():
    # A ideia dessa função é com base em varias desacelerações imputadas pelo usuario a resposta seja os pesos aparentes.
    # Tentar plotar gráficos se possível
    # Material do Erthal, usar a ideia do gráfico da pagina 21 com inputs sendo as acelerações do usuário.
    # trabalhar com um range consideravel.
    #
    pass

def IDEAL_Ffront_Frear():
    # Aqui o objetivo é plotar o grafico (x,y)=Ffront vs Frear com uma reta meio curva passando após achar o ideal...
    # Usar o gráfico do material do Erthal, página 24.
    # Fazer duas situações, pista molhada e pista seca (Dois gráficos)
    # Também fazer os quadrantes separando "trava tal, trava outra, trava ambas"...
    pass
# -------------------- Funções Suporte ----------------------------------

def tit_txt():
    prototipo = str(input("Nome do Protótipo: ")).strip().upper()
    evic = str(input("EV ou IC? ")).strip().upper()
    data_atual = datetime.now()
    data = data_atual.strftime("%d/%m/%Y")
    hora = data_atual.strftime("%H:%M:%S")

    # Exibe a data formatada
    print("==========================================================================================================")
    print(f"PARÂMETROS DO PROTOTIPO: {prototipo}")
    print(f"MODALIDADE: {evic}")
    print(f"Dados Gerados {data} as {hora}.")
    print("==========================================================================================================")
    print("Força Aplicada no Pedal------------- | 2000N")
    print("Ganho Mecânico---------------------- | 3,6")
    print("BIAS (FRONT/REAR) ------------------ | 62%/48%")
    print("Pressão (FRONT/REAR) --------------- | 28 Bar/17 Bar")
    print("")


```
